```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
```

```python
video_df = pd.read_csv("videoresults1.csv")
video_df.head()
```

```python
plt.figure(figsize = (15, 8))
plt.plot(video_df['Total Vehicals'], label='Number of Vehicals in both lanes', color='blue', linewidth = 1)
plt.plot(video_df['Number of Vehicals in Left'], label='Number of cars in left lanes', color='purple', linewidth = 1)
plt.plot(video_df['Number of Vehicals in Right'], label='Number of cars in right lanes', color='red', linewidth = 1)
plt.yticks(np.arange(0 , max(video_df['Total Vehicals']) , 3))
plt.legend()
plt.grid(color='gray', linestyle='-', linewidth=0.5)
plt.xlabel('Frames', fontsize = 12)
plt.ylabel('num of vehicals', fontsize = 12)
plt.title('Number of detected vehicals by lanes', fontsize = 20 , fontweight = "bold")

plt.savefig("number_of_vehicals_plot.png", dpi=250)
plt.show()
```

```python
plt.hist(video_df['Total Vehicals'] ,bins = 5 ,  ec = "red")
plt.xlabel("Number vehicals")
plt.ylabel("count")
plt.title("Histogram")
plt.savefig("Histogram.jpg" , dpi = 250)
```

```python
ax = video_df["Traffic in Left"].value_counts().plot(kind="bar")

plt.title("Left lane Traffic Status")
ax.set_xlabel("Left lane Traffic Status")
ax.set_ylabel("Numbers")

plt.savefig("left_lane_traffic_state.jpg" , dpi = 250)
```

```python
ax = video_df["Traffic in Right"].value_counts().plot(kind="bar")

plt.title("Right lane Traffic Status")
ax.set_xlabel("Right lane Traffic Status")
ax.set_ylabel("Numbers")

plt.savefig("right_lane_traffic_state.jpg" , dpi = 250)
```
